## Run this code to get your dataset downloaded and the path.

```python
import kagglehub
import os
import shutil
target_folder = os.path.abspath("../dataset")
if os.path.exists(target_folder) and os.listdir(target_folder):
    print("Data already exists")
else:
    os.makedirs(target_folder, exist_ok=True)
    
    path = kagglehub.dataset_download("orvile/gastric-cancer-histopathology-tissue-image-dataset")
    print("Path to dataset files:", path)
    
    shutil.move(path, target_folder)
    print("Dataset moved to:", target_folder)
```
